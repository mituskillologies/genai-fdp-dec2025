```python
import chromadb
```

```python
chroma_client = chromadb.Client()
```

```python
collection = chroma_client.create_collection(name="my_collection")
```

```python
collection.add(
    ids = ["id1", "id2"],
    documents= [
        "tomato is a fruit",
        "apple is a brand"
    ]
)
```

```python
collection.get()
```

```python
# query the collection
```

```python
result = collection.query(
    query_texts= ["company"],
    n_results= 2
)
```

```python
result
```

```python
collection.upsert(
    ids = ["id2"],
    documents= [
        "king arthur drew his sword"
    ]
)
```

```python
collection.get()
```

```python
result = collection.query(
    query_texts= ["royalty"],
    n_results= 2
)
```
