```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2', pad_token_id=tokenizer.eos_token_id)
```

```python
prompt = "the future of AI is"
input_ids = tokenizer.encode(prompt, return_tensors='pt')
```

```python
# greedy
```

```python
print(tokenizer.decode(model.generate(input_ids, max_length=100,
                                      do_sample=False)[0]))
```

```python
# beam search
```

```python
print(tokenizer.decode(model.generate(input_ids, max_length=50,
                                      num_beams = 5)[0]))
```

```python
# top-k
```

```python
print(tokenizer.decode(model.generate(input_ids, max_length=50,
                                      do_sample = True,
                                      top_k = 15)[0]))
```

```python
# top-p
print(tokenizer.decode(model.generate(input_ids, max_length=50,
                                      do_sample = True,
                                      top_p = 0.7)[0]))
```
